```python
import numpy as np
import pandas as pd
import seaborn as sb
df=pd.read_csv('heart.csv')
```

## Correlation

```python
df.corr()
```

## HEATMAP

```python
sb.heatmap(df.corr())
```

```python
sb.jointplot(x='Age',y='MaxHR', data=df)
```

```python
sb.jointplot(x='Age',y='MaxHR', data=df,kind='hex')
```

## Regression line shows that with increasing age MaxHR decreases

```python
sb.jointplot(x='Age',y='MaxHR', data=df,kind='reg')
```

## No Correlation

```python
sb.jointplot(y='Oldpeak',x='HeartDisease', data=df)
```

## Pair Plot - tells Correlation between just two columns

```python
sb.pairplot(df)
```
